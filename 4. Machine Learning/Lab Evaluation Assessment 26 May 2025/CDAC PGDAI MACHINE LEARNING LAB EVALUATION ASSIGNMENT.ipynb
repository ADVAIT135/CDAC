## 2. Aggregating Data to Monthly Sales

```python
monthly_agg = df.groupby('date_block_num').agg({
    'item_cnt_day': 'sum',      # Total sales per month
    'item_price': 'mean'        # Average price per month
}).reset_index()
```

```python
monthly_agg.rename(columns={'item_cnt_day': 'total_sales', 
                            'item_price': 'avg_price'}, inplace=True)
```

## 4. Outlier Detection using IQR Method on Monthly Sales

```python
Q1 = monthly_agg['total_sales'].quantile(0.25)
Q3 = monthly_agg['total_sales'].quantile(0.75)
IQR = Q3 - Q1
outlier_condition = (monthly_agg['total_sales'] < Q1 - 1.5 * IQR) | (monthly_agg['total_sales'] > Q3 + 1.5 * IQR)
outliers = monthly_agg[outlier_condition]

print("Outlier months based on total sales:")
print(outliers)
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(monthly_agg['total_sales'], vert=False)
plt.title('Boxplot of Monthly Total Sales')
plt.xlabel('Total Sales')
plt.show()
```

## 5. Creating Lag Features and Rolling Means

```python
# Create lag features (e.g., 1-month and 2-month lags for total sales)
monthly_agg['sales_lag1'] = monthly_agg['total_sales'].shift(1)
monthly_agg['sales_lag2'] = monthly_agg['total_sales'].shift(2)
```

```python
# Compute a 3-month rolling mean for smoothing
monthly_agg['rolling_mean_3'] = monthly_agg['total_sales'].rolling(window=3).mean()
```

```python
# Calculate month-over-month percentage change in average price
monthly_agg['price_pct_change'] = monthly_agg['avg_price'].pct_change()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(monthly_agg['date_block_num'], monthly_agg['avg_price'], marker='o', label='Avg Price')
plt.plot(monthly_agg['date_block_num'], monthly_agg['price_pct_change'] * 100, marker='x', label='Price % Change')
plt.title('Monthly Average Price and Price Percentage Change')
plt.xlabel('Month (date_block_num)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45, ha='right')

# Adjust the layout to accommodate the rotated labels
plt.tight_layout()
plt.legend()
plt.show()
```

## 6. XGBoost for Time Series Forecasting

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
data = monthly_agg.dropna().reset_index(drop=True)
```

```python
features = ['sales_lag1', 'sales_lag2', 'rolling_mean_3', 'price_pct_change']
target = 'total_sales'
```

```python
train_size = int(len(data) * 0.8)
X_train = data[features].iloc[:train_size]
y_train = data[target].iloc[:train_size]
X_test = data[features].iloc[train_size:]
y_test = data[target].iloc[train_size:]
```

```python
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
xgb_model.fit(X_train, y_train)
```

```python
y_pred = xgb_model.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
print("XGBoost Model Performance:")
print(f"RMSE: {rmse:.2f}")
print(f"MAE:  {mae:.2f}")
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data['date_block_num'].iloc[train_size:], y_test, marker='o', label="Actual Sales")
plt.plot(data['date_block_num'].iloc[train_size:], y_pred, marker='x', label="Predicted Sales")
plt.xlabel('Month (date_block_num)')
plt.ylabel('Total Sales')
plt.title('XGBoost: Actual vs Predicted Sales')
plt.legend()
plt.show()
```

## 7. LSTM for Time Series Forecasting

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sales_values = data['total_sales'].values.reshape(-1, 1)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
sales_scaled = scaler.fit_transform(sales_values)
```

```python
def create_sequences(series, window_size):
    X, y = [], []
    for i in range(len(series) - window_size):
        X.append(series[i:i + window_size])
        y.append(series[i + window_size])
    return np.array(X), np.array(y)
```

```python
window_size = 3
X, y = create_sequences(sales_scaled, window_size)
```

```python
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
print("LSTM input shape:", X_train.shape)
```

```python
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(window_size, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
history = model.fit(X_train, y_train, epochs=50, batch_size=1, verbose=1)
```

```python
y_pred_lstm = model.predict(X_test)
```

```python
y_pred_inv = scaler.inverse_transform(y_pred_lstm)
y_test_inv = scaler.inverse_transform(y_test)
```

```python
rmse_lstm = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
mae_lstm = mean_absolute_error(y_test_inv, y_pred_inv)
print("LSTM Model Performance:")
print(f"RMSE: {rmse_lstm:.2f}")
print(f"MAE:  {mae_lstm:.2f}")
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(len(y_test_inv)), y_test_inv, marker='o', label="Actual Sales")
plt.plot(range(len(y_pred_inv)), y_pred_inv, marker='x', label="Predicted Sales")
plt.xlabel('Time Step')
plt.ylabel('Total Sales')
plt.title('LSTM: Actual vs Predicted Sales')
plt.legend()
plt.show()
```

## 8. Forecasting next month data using the XGB Model

```python
df = merged_df_3
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')
```

```python
df['date_block_num'] = (df['date'].dt.year - 2013) * 12 + (df['date'].dt.month - 1)
```

```python
monthly_sales = (
    df.groupby(['date_block_num', 'shop_id', 'item_id'], as_index=False)
      ['item_cnt_day'].sum()
      .rename(columns={'item_cnt_day': 'monthly_sales'})
)
```

```python
all_months = np.arange(monthly_sales['date_block_num'].min(), monthly_sales['date_block_num'].max() + 1)
all_shops = monthly_sales['shop_id'].unique()
all_items = monthly_sales['item_id'].unique()
```

```python
full_grid = pd.MultiIndex.from_product(
    [all_months, all_shops, all_items],
    names=['date_block_num', 'shop_id', 'item_id']
).to_frame(index=False)
```

```python
monthly_sales = full_grid.merge(monthly_sales, on=['date_block_num', 'shop_id', 'item_id'], how='left')
monthly_sales['monthly_sales'] = monthly_sales['monthly_sales'].fillna(0)
```

```python
def create_lag_features(df, lags, col):
    """
    For each lag in "lags", create a new column with shifted values of `col`.
    """
    for lag in lags:
        df[f'{col}_lag_{lag}'] = df.groupby(['shop_id', 'item_id'])[col].shift(lag)
    return df
```

```python
lags = [1, 2, 3]
monthly_sales = create_lag_features(monthly_sales, lags, 'monthly_sales')
```

```python
monthly_sales['rolling_mean_3'] = monthly_sales.groupby(['shop_id', 'item_id'])['monthly_sales'] \
    .transform(lambda x: x.shift(1).rolling(window=3, min_periods=1).mean())
```

```python
monthly_sales.fillna(0, inplace=True)
```

```python
max_month = monthly_sales['date_block_num'].max()
train = monthly_sales[monthly_sales['date_block_num'] < max_month]
```

```python
test_grid = pd.MultiIndex.from_product(
    [[max_month + 1], all_shops, all_items],
    names=['date_block_num', 'shop_id', 'item_id']
).to_frame(index=False)
```

```python
last_month_features = monthly_sales[monthly_sales['date_block_num'] == max_month][[
    'shop_id', 'item_id', 'monthly_sales', 'monthly_sales_lag_1', 'monthly_sales_lag_2', 'monthly_sales_lag_3', 'rolling_mean_3'
]]
```

```python
test = test_grid.merge(last_month_features, on=['shop_id', 'item_id'], how='left')
test.fillna(0, inplace=True)
```

```python
feature_columns = ['monthly_sales_lag_1', 'monthly_sales_lag_2', 'monthly_sales_lag_3', 'rolling_mean_3']
target_column = 'monthly_sales'
```

```python
X_train = train[feature_columns]
y_train = train[target_column]
```

```python
model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42, n_estimators=100)
model.fit(X_train, y_train)
```

```python
train_preds = model.predict(X_train)
rmse = np.sqrt(mean_squared_error(y_train, train_preds))
print("Training RMSE: {:.2f}".format(rmse))
```

```python
X_test = test[feature_columns]
test['predicted_sales'] = model.predict(X_test)
```
