```python
def ex_2():
  user_input = input('Enter the input in YYYY-MM-DD HH:MM:SS format -')
  try:
    given_date = datetime.datetime.strptime(user_input, '%Y-%m-%d %H:%M:%S')
    print('Given Date : ', given_date)
    next_date = given_date + datetime.timedelta(days=7, hours = 12)
    print('Next Date : ', next_date)

  except ValueError:
    print("Invalid date format. Please use YYYY-MM-DD HH:MM:SS.")
```

```python
def ex_4():
  date_1_str = input('Enter the 1st date in yyyy-mm-dd format : ')
  date_2_str = input('Enter the 2nd date in yyyy-mm-dd format : ')
  try:
        date_1 = datetime.datetime.strptime(date_1_str, '%Y-%m-%d').date()
        date_2 = datetime.datetime.strptime(date_2_str, '%Y-%m-%d').date()

        print('The First Date is:', date_1)
        print('The Second Date is:', date_2)

        days_diff = (date_2 - date_1).days

        print(f'The days between {date_1} and {date_2} is {days_diff} days')
  except ValueError:
        print("Invalid date format. Please use YYYY-MM-DD.")
```

```python
def ex_5():


    now = datetime.datetime.now()

    # a) Current date and time
    print("Current date and time:", now)

    # b) Current year in full
    print("Current year:", now.strftime("%Y"))

    # c) Month of year full name
    print("Month of year:", now.strftime("%B"))

    # d) Weekday of the week (0=Monday, 6=Sunday)
    print("Weekday of the week (number):", now.weekday())

    # e) Day of the month
    print("Day of the month:", now.strftime("%d"))

    # f) Day of week in full name
    print("Day of week (full name):", now.strftime("%A"))
```

```python
class Triangle:
    number_of_sides = 3  # Class variable for the number of sides

    def __init__(self, angle1, angle2, angle3):
        """Initializes a Triangle object with the given angles."""
        self.angle1 = angle1
        self.angle2 = angle2
        self.angle3 = angle3

    def check_angles(self):
        """Checks if the sum of angles is 180 degrees."""
        return self.angle1 + self.angle2 + self.angle3 == 180

# Create an instance of the Triangle class
my_triangle = Triangle(90, 30, 60)

# Print the number of sides and check_angles result
print(my_triangle.number_of_sides)  # Output: 3
print(my_triangle.check_angles())   # Output: True
```

```python
my_triangle = Triangle(45, 15, 120)
print(my_triangle.number_of_sides)
print(my_triangle.check_angles())
```

```python
class Songs:
    def __init__(self, lyrics):
        """Initializes a Songs object with the given lyrics."""
        self.lyrics = lyrics

    def sing_me_a_song(self):
        """Prints each line of the lyrics on a separate line."""
        for line in self.lyrics:
            print(line)

# Create an instance of the Songs class
happy_bday = Songs(["May god bless you, ",
                    "Have a sunshine on you,",
                    "Happy Birthday to you !"])

# Call the sing_me_a_song method
happy_bday.sing_me_a_song()
```

```python
class Lunch:
    def __init__(self, menu):
        self.menu = menu

    def menu_price(self):
        if self.menu == "menu 1":
            print("Your choice:", self.menu, "Price 12.00")
        elif self.menu == "menu 2":
            print("Your choice:", self.menu, "Price 13.40")
        else:
            print("Error in menu")

```

```python
class ex_10:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth

    def perimeter(self):
        return 2 * (self.length + self.breadth)


# Taking input from the user
length = int(input("Enter Length of a Rectangle: "))
breadth = int(input("Enter Breadth of a Rectangle: "))

# Creating an object of the Rectangle class
rect = ex_10(length, breadth)

while True:
    print("\n1. Area")
    print("2. Perimeter")
    print("3. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        print(f"Area of rectangle with length {length} and breadth {breadth} is {rect.area()}")
    elif choice == 2:
        print(f"Perimeter of rectangle with length {length} and breadth {breadth} is {rect.perimeter()}")
    elif choice == 3:
        print("End of the Program")
        break
    else:
        print("Invalid choice! Please select again.")
```
