```python
def add_suffix(s):
    if len(s) >= 3:
        if s.endswith("ing"):
            return s + "ly"
        else:
            return s + "ing"
    return s
```

```python
def student_percent_division():
    print("Enter the marks of student obtained in each of the subjects out of 100: -\n")
    sub_1 = int(input("1. English : "))
    sub_2 = int(input("\n2. Marathi : "))
    sub_3 = int(input("\n3. Sanskrit : "))
    sub_4 = int(input("\n4. Science and Technology : "))
    sub_5 = int(input("\n5. Mathematics : \n"))
    total_marks = sub_1 + sub_2 + sub_3 + sub_4 + sub_5
    print("The total marks the student obtained is", total_marks, "out of 500.\n")
    percentage_score = ((total_marks/500) * 100)
    print("The student has scored a overall percentage of : ", percentage_score, "%")
    if percentage_score >= 60:
        division = "First Division"
    elif percentage_score >= 50 and percentage_score <= 59:
        division = "Second Division"
    elif percentage_score >= 40 and percentage_score <= 49:
        division = "Third Division"
    elif percentage_score < 40:
        print("The student has failed the Semester he/she needs to reappear!!!!!")
    if percentage_score > 40:
        print("Congratulations!!! You have passed this Semester with", percentage_score, "%. Securing a ",division)
    
    
```

```python
def leap_year():
    year = int(input("Enter the year : "))
    if year % 4 == 0 or year % 100 == 0 and year % 400 == 0:
        print("The entered year ", year, "is a leap year!!!")
    else:
        print("The entered year ", year, "is not a leap year!!!")
```

```python
def palindrome_string_checker():
    str_a = input("Enter any string: ")
    str_b = str_a[::-1]
    if str_a == str_b:
        print("The entered string ", str_a, 'is a palidrome string!!!')
    else:
        print("The entered string ", str_a, 'is not a palidrome string!!!')
```
