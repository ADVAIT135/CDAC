```python
def ex_24():
  num = int(input("Enter any number : "))
  sum = 0
  for i in range(2,num):
    if num%i == 0:
      sum = sum + 1
    else:
      sum = sum + 0
  if sum == 0:
    print(f'The entered number {num} is a prime number!!!')
  else:
    print(f'The entered number {num} is a not prime number!!!')
```

```python
def ex_25():
  num = int(input("Enter the count of numbers you want to have in the list: "))
  list_num = []
  for i in range(0,num):
    element = int(input(f'\nEnter the value of {i} number: '))
    list_num.append(element)
  print(f'The list entered by you is : {list_num}')
  pos_list = []
  neg_list = []
  for i in list_num:
    if i > 0:
      pos_list.append(i)
    else:
      neg_list.append(i)
  print(f'\nThe list of positive numbers from {list_num} is {pos_list} and it\'s sum is {sum(pos_list)}.')
  print(f'\nThe list of negative numbers from {list_num} is {neg_list} and it\'s sum is {sum(neg_list)}.')
```
