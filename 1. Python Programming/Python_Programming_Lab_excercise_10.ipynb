```python
class InvalidAgeError(Exception):
    def __init__(self, age, message="Age must be between 0 and 150"):
        self.age = age
        self.message = message
        super().__init__(f"InvalidAgeError: {message}. Given age: {age}")

def get_age():
    try:
        age = int(input("Enter your age: "))
        if age < 0 or age > 150:
            raise InvalidAgeError(age)
        print(f"Valid age entered: {age}")
    except InvalidAgeError as e:
        print(e)
    except ValueError:
        print("Invalid input! Please enter a valid integer for age.")

```

```python
def inner_function():

    raise ValueError("An error occurred in inner_function.")

def outer_function():

    try:
        inner_function()
    except ValueError as e:
        raise RuntimeError("Outer function encountered an error while calling inner_function.") from e
```

```python
try:
    outer_function()
except RuntimeError as e:
    print(f"Caught Exception: {e}")
    print(f"Original Exception: {e.__cause__}")
```

```python
def ex_10_9(filename):
    try:
        with open(filename, "r") as file:
            for line in file:
                try:
                    number = int(line.strip())  # Attempt to convert line to an integer
                    print(f"Processed number: {number}")
                except ValueError:
                    print(f"Skipping invalid line: {line.strip()}")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")

filename = "numbers.txt"
ex_10_9(filename)
```

```python
def check_integer(x):
    if not isinstance(x, int):
        raise TypeError("Error: x must be an integer!")
    print(f"x is a valid integer: {x}")
```

```python
def ex_10_12():
  try:
    x = input('Enter anything!!! : ')
    x = int(x)
    check_integer(x)
  except ValueError:
      raise TypeError("Error: x must be an integer!")
```
