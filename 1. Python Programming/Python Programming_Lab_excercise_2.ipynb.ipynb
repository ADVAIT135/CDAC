```python
def replace_char(s):
    first_char = s[0]  
    modified_str = first_char + s[1:].replace(first_char, '$') 
    return modified_str

sample_str = 'restart'
result = replace_char(sample_str)
print(result)
```

```python
sample_str = 'Advait'
result = replace_char(sample_str)
print(result)
```

```python
def swap_first_two(str1, str2):
    
    new_str1 = str2[:2] + str1[2:]
    new_str2 = str1[:2] + str2[2:]
    return new_str1 + " " + new_str2


str1 = 'abc'
str2 = 'xyz'

result = swap_first_two(str1, str2)
print(result)
```
