```python
def count_vowels():
    s = input("Enter any String")
    vowels = "aeiou"
    vowel_count = {v: 0 for v in vowels} 

    for char in s.lower():
        if char in vowel_count:
            vowel_count[char] += 1

    vowel_count = {k: v for k, v in vowel_count.items() if v > 0}

    print(vowel_count)
```

```python
count_vowels()
```

```python
def sort_string():
    str = input("Enter any string: ")
    str_list = str.split()
    str_list.sort()
    print("The sorted string list for the given string is ", " ".join(str_list))
        
```

```python
sort_string()
```

```python
import string

def remove_punctuation():
    user_input = input("Enter a String :")
    print("".join(char for char in user_input if char not in string.punctuation))
```

```python
remove_punctuation()
```

```python
def triangle(n):
    # Upper part of the triangle
    for i in range(1, n + 1):
        print("* " * i)
    
    # Lower part of the triangle
    for i in range(n - 1, 0, -1):
        print("* " * i)

triangle(4)
```

```python
def multiplication_table():
    for i in range(1, 11):  # Rows
        for j in range(1, 11):  # Columns
            print(f"{i * j:4}", end=" ")  # Print with spacing
        print()  # New line after each row
```

```python
multiplication_table()
```
