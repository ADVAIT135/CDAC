```python
def read_num_lines():
    file_name = input("Enter the file name : ")
    f = open(file_name, encoding = 'utf-8')
    line_nums = int(input("Enter the number of lines you wanna read from the file :"))
    for i in range(line_nums):
        print(f.readline())
    f.close()
```

```python
def write_to_file():
    file_name = input("Enter the file name you want to add lines in: ")

    with open(file_name, 'a+', encoding='utf-8') as f:
        f.seek(0)  # Move cursor to beginning
        print("The contents of the file before the write operations are:\n")
        print(f.read())  

        line_nums = int(input("\nEnter the number of lines you want to add in file: "))
        
        for i in range(line_nums):
            line = input(f"Enter line {i+1}: ")
            f.write("\n" + line)  
        

        f.seek(0)
        print("\nThe contents of the file after the write operations are:\n")
        print(f.read())
```

```python
write_to_file()
```

```python
def print_lines_in_reverse():
    file_name = input("Enter the name of the file you want to read : ")
    with open(file_name, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        print("The contents of file before reverse", lines)
    print("\nThe contents of file after reverse :-")
    for line in lines:
        print(line.strip()[::-1]) 
```

```python
print_lines_in_reverse()
```

```python
def list_to_file():
    num_list = int(input("Enter the number of elements you want to have in the list: "))
    element_list = []
    for i in range(num_list):
        element = input(f"\nEnter the {i+1} element of the list : ")
        element_list.append(element)
    print("\nThe list is :", element_list)
    file_name = input("Enter the filename you want to add the list elements in : ")
    f = open(file_name, 'a+', encoding = 'utf-8')
    f.seek(0)
    print(f"The file before adding the list elements into it is: -" )
    print(f.read())
    for i in element_list:
        f.write(i + "\n")
    f.seek(0)
    print(f"The file after adding the list elements into it is : -")
    print(f.read())
```

```python
list_to_file()
```

```python
def count_file_contents():
    file_name = input("Enter the file name: ")
    try: 
        f = open(file_name, encoding = 'utf-8')
        content = f.readlines()
        num_lines = len(content)
        num_words = sum(len(line.split()) for line in content)
        num_chars = sum(len(line) for line in content)
        print("The number of lines in the file are : ", num_lines)
        print("The number of words in the file are : ", num_words)
        print("The number of characters in the file are: ", num_chars)
    except FileNotFoundError:
        print("Error: The file does not exist.")
        
```

```python
def ex_7_7():
    file_name = input("Enter the file name you want to save the contents in : ")
    with open(file_name, 'w', encoding='utf-8') as file:
        print("Enter sentences (Type 'END' to stop):")
        while True:
            sentence = input()
            if sentence == "END":
                break
            file.write(sentence + "\n")
    file.close()
    with open(file_name, 'r', encoding='utf-8') as file:
        print("\nSentences starting with an uppercase letter:")
        for line in file:
            if line.strip() and line[0].isupper():
                print(line.strip())
    file.close()
```

```python
ex_7_7()
```

```python
import pickle
def ex_7_8():
    file_name = input("Enter the name of the file with which you want to save the file : ")
    num_records = int(input("Enter the number of records to be added: "))
    
    with open(file_name, 'wb') as file:
        records = []
        for _ in range(num_records):
            item_no = int(input("Enter Item No: "))
            item_name = input("Enter Item Name: ")
            qty = int(input("Enter Quantity: "))
            price = float(input("Enter Price per item: "))
            
            
            records.append((item_no, item_name, qty, price))
        
        
        pickle.dump(records, file)
    file.close()
    try:
        with open(file_name, 'rb') as file:
            records = pickle.load(file)
            
            print("\nRecords in the file:\n")
            for record in records:
                item_no, item_name, qty, price = record
                amount = qty * price
                print(f"Item No: {item_no}")
                print(f"Item Name: {item_name}")
                print(f"Quantity: {qty}")
                print(f"Price per item: {price}")
                print(f"Amount: {amount}\n")
    
    except FileNotFoundError:
        print("File not found. Please write data first.")
```
