```python
def add(a, b):
    print(a, "+", b, "=", a + b)

def sub(a, b):
    print(a, "-", b, "=", a - b)

def mul(a, b):
    print(a, "*", b, "=", a * b)

def div(a, b):
    if b == 0:
        print("Error! Division by zero is not allowed.")
    else:
        print(a, "/", b, "=", a / b)

def simple_calculator():
    while True:
        print("\nSelect operation:")
        print("1. Add")
        print("2. Subtract")
        print("3. Multiply")
        print("4. Divide")
        print("5. Exit")

        choice = int(input("Enter choice (1/2/3/4/5): "))

        if choice == 5:
            print("Exiting the calculator. Goodbye!")
            break

        a = float(input("Enter First Number: "))
        b = float(input("Enter Second Number: "))

        if choice == 1:
            add(a, b)
        elif choice == 2:
            sub(a, b)
        elif choice == 3:
            mul(a, b)
        elif choice == 4:
            div(a, b)
        else:
            print("Invalid choice! Please enter a valid option.")
```

```python
simple_calculator()
```

```python
def fibonacci():
    n = int(input("Enter the number of terms: "))
    if n <= 0:
        print("Please enter a positive integer.")
    else:
        fib_sequence = [0, 1]
        for i in range(2, n):
            next_term = fib_sequence[-1] + fib_sequence[-2]
            fib_sequence.append(next_term)
        print("Fibonacci sequence up to", n, "terms:", fib_sequence[:n])

```

```python
def num_div_13():
    num_terms = int(input("Enter the number of terms you want to verify whether they are divisible by 13 or not : "))
    list_terms = []
    for i in range(num_terms):
        x = int(input(f"Enter the {i+1} term."))
        list_terms.append(x)
    numbers_div_by_13 = list(filter(lambda x : x % 13 == 0, list_terms))
    print("The numbers from the list",list_terms,"that are divisible by 13 are",numbers_div_by_13)
```

```python
def fibonacci(n):
    if n <= 0:
        return "Invalid input! Enter a positive integer."
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
n_terms = int(input("Enter the number of terms: "))

print("Fibonacci sequence:")
for i in range(1, n_terms + 1):
    print(fibonacci(i), end=" ")
```

```python
def sum_natural_num(n):
    if n <= 0:
        return 0
    else:
        return n + sum_natural_num(n-1)

n = int(input("Enter any number : "))

if n < 0:
    print("Please enter a Positive Number. The number you have entered is less than 0 which is not a natural number!!!!")
else:
    print("The sum of",n,"natural numbers is:", sum_natural_num(n))
    
```

```python
def palidrome_recognizer():
    str_a = input("Enter the phrase that you want to verify whether it's a palidrome or not : ")
    str_b = []
    for i in str_a.lower():
        if i.isalpha():
            str_b.append(i)
    palidrome_str_b = str_b[::-1]
    if str_b == palidrome_str_b:
        print(f"The entered string'{str_a}'is a palidrome string")
    else:
        print(f"The entered string'{str_a}'is not a palidrome string")
```

```python
import string

def panagram():
    str_a = input("Enter the string that you want to verify whether it's a Panagram or not: ")
    str_a = str_a.lower()
    letters_str_a = set(str_a)
    if set(string.ascii_lowercase).issubset(letters_str_a):
        print(f"The entered string '{str_a}' is a Panagram string.")
    else:
        print(f"The entered string '{str_a}' is not a Panagram string.")
```

```python
def overlapping():
    list_a = []
    list_b = []
    num_a, num_b = map(int,input("Enter the number of elements you want to have in the two list : ").split())
    for i in range(num_a):
        x = input(f"Enter the {i+1} element of the 1st List: ")
        list_a.append(x)
    for i in range(num_b):
        y = input(f"Enter the {i+1} element of the 2nd List: ")
        list_b.append(y)
    print("\n The two Lists are: - ")
    print("\n List A:", list_a)
    print("\n List B:", list_b)
    common_elements = list(set(list_a) & set(list_b))
    if len(common_elements) > 0:
        print("The common elements present in both list A and B are: ", common_elements)
        return True
    else:
        print("There are no common elements present in both of the list")
        return False
        
```

```python
def find_longest_word():
    list_num = int(input("Enter the number of words you want to have in the list : "))
    list_words = []
    for i in range(list_num):
        word = input(f"Enter the {i+1} word of the List")
        list_words.append(word)
    print(f"The length of the longest word is {max(len(word) for word in list_words)}")
```

```python
def filter_long_words():
    list_num = int(input("Enter the number of words you want to have in the list : "))
    list_words = []
    for i in range(list_num):
        word = input(f"Enter the {i+1} word of the List")
        list_words.append(word)
    n = int(input("Enter any number : "))
    print("The list of words whose length is longer than",n,"are",list(word for word in list_words if len(word) > n))
```
