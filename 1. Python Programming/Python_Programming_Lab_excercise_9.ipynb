```python
class BankAccount:
    def __init__(self, name, acc_number, acc_type, balance=0.0):
        """Assign initial values."""
        self.name = name
        self.acc_number = acc_number
        self.acc_type = acc_type
        self.balance = balance

    def deposit(self, amount):
        """Deposit an amount into the account."""
        if amount > 0:
            self.balance += amount
            print(f"{amount} deposited successfully. New balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        """Withdraw an amount from the account."""
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"{amount} withdrawn successfully. New balance: {self.balance}")
        else:
            print("Insufficient balance or invalid amount.")

    def display(self):
        """Display the account holder's name and balance."""
        print("\nAccount Details:")
        print(f"Account Holder: {self.name}")
        print(f"Account Number: {self.acc_number}")
        print(f"Account Type: {self.acc_type}")
        print(f"Balance: {self.balance}\n")
```

```python
def ex_9_2():
  # Menu-driven program
  account = None

  while True:
      print("\n1. New customer")
      print("2. Deposit")
      print("3. Withdrawal")
      print("4. Display")
      print("5. Exit")

      choice = input("Enter your choice: ")

      if choice == "1":
          name = input("Enter Name of Depositor: ")
          acc_number = input("Enter Account Number: ")
          acc_type = input("Enter Account Type (Savings/Current): ")
          balance = float(input("Enter Initial Balance: "))
          account = BankAccount(name, acc_number, acc_type, balance)
          print("Account created successfully!")

      elif choice == "2":
          if account:
              amount = float(input("Enter deposit amount: "))
              account.deposit(amount)
          else:
              print("No account found! Please create an account first.")

      elif choice == "3":
          if account:
              amount = float(input("Enter withdrawal amount: "))
              account.withdraw(amount)
          else:
              print("No account found! Please create an account first.")

      elif choice == "4":
          if account:
              account.display()
          else:
              print("No account found! Please create an account first.")

      elif choice == "5":
          print("Exiting... Thank you!")
          break

      else:
          print("Invalid choice! Please enter a number between 1 and 5.")
```

```python
class Staff:
    """Base class for all staff members."""
    def __init__(self, code, name):
        self.code = code
        self.name = name

    def display(self):
        print(f"Code: {self.code}, Name: {self.name}")

class Teacher(Staff):
    """Derived class for teachers with subject and publication."""
    def __init__(self, code, name, subject, publication):
        super().__init__(code, name)
        self.subject = subject
        self.publication = publication

    def display(self):
        super().display()
        print(f"Subject: {self.subject}, Publication: {self.publication}")

class Typist(Staff):
    """Base class for typists with typing speed."""
    def __init__(self, code, name, speed):
        super().__init__(code, name)
        self.speed = speed

    def display(self):
        super().display()
        print(f"Typing Speed: {self.speed} WPM")

class Regular(Typist):
    """Derived class for regular typists with salary."""
    def __init__(self, code, name, speed, salary):
        super().__init__(code, name, speed)
        self.salary = salary

    def display(self):
        super().display()
        print(f"Salary: {self.salary}")

class Casual(Typist):
    """Derived class for casual typists with daily wages."""
    def __init__(self, code, name, speed, daily_wages):
        super().__init__(code, name, speed)
        self.daily_wages = daily_wages

    def display(self):
        super().display()
        print(f"Daily Wages: {self.daily_wages}")

class Officer(Staff):
    """Derived class for officers with grade."""
    def __init__(self, code, name, grade):
        super().__init__(code, name)
        self.grade = grade

    def display(self):
        super().display()
        print(f"Grade: {self.grade}")
```

```python
def ex_9_2_2():

  # Menu-driven program
  staff_list = []

  while True:
      print("\n1. Teacher")
      print("2. Officer")
      print("3. Regular Typist")
      print("4. Casual Typist")
      print("5. Exit")

      choice = input("Enter your choice: ")

      if choice == "1":
          code = input("Enter Code: ")
          name = input("Enter Name: ")
          subject = input("Enter Subject: ")
          publication = input("Enter Publication: ")
          staff_list.append(Teacher(code, name, subject, publication))
          print("Teacher added successfully!")

      elif choice == "2":
          code = input("Enter Code: ")
          name = input("Enter Name: ")
          grade = input("Enter Grade: ")
          staff_list.append(Officer(code, name, grade))
          print("Officer added successfully!")

      elif choice == "3":
          code = input("Enter Code: ")
          name = input("Enter Name: ")
          speed = input("Enter Typing Speed (WPM): ")
          salary = float(input("Enter Salary: "))
          staff_list.append(Regular(code, name, speed, salary))
          print("Regular Typist added successfully!")

      elif choice == "4":
          code = input("Enter Code: ")
          name = input("Enter Name: ")
          speed = input("Enter Typing Speed (WPM): ")
          daily_wages = float(input("Enter Daily Wages: "))
          staff_list.append(Casual(code, name, speed, daily_wages))
          print("Casual Typist added successfully!")

      elif choice == "5":
          print("\nStaff Details:")
          for staff in staff_list:
              staff.display()
              print("-------------------")
          print("Exiting... Thank you!")
          break

      else:
          print("Invalid choice! Please enter a number between 1 and 5.")
```
