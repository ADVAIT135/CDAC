## 2. Pokemon details as per the user input

```python
def user_input_pokemon_name():
    pokemon_name = input("Enter a Pokemon name: ").strip()

    selected = df[df['name'].str.lower() == pokemon_name.lower()]
    
    if selected.empty:
        print(f"No Pokémon found with the name '{pokemon_name}'.")
    else:
        print(f"Details for {pokemon_name}:")
        print(selected)
```

## 3. Pokemon with maximum Total

```python
max_total = df['base_total'].max()

candidates = df[df['base_total'] == max_total]

# If there's more than one candidate, narrow it down to those that are Legendary.
if len(candidates) > 1:
    candidates_legend = candidates[candidates['is_legendary'] == True]
    if not candidates_legend.empty:
        candidates = candidates_legend

# If still more than one candidate, choose those from Generation 3.
if len(candidates) > 1:
    candidates_gen3 = candidates[candidates['generation'] == 3]
    if not candidates_gen3.empty:
        candidates = candidates_gen3

print("\nPokemon with the maximum Total (after applying tie-breakers): \n")
print(candidates['name'])
print(candidates)
```

```python
candidates.to_excel('3. Pokemon with maximum total.xlsx', index = False)
```
