```python
missing_values = ["N.A."]
```

```python
df_2 = pd.read_excel('fortune500.xlsx', na_values=missing_values)
```

```python
print("Missing data count:")
print(df_2.isnull().sum())
```

```python
df_2 = df_2.dropna(subset=['Revenue (in millions)', 'Profit (in millions)'])
```

```python
df_2['Revenue (in millions)'] = pd.to_numeric(df_2['Revenue (in millions)'], errors='coerce')
df_2['Profit (in millions)'] = pd.to_numeric(df_2['Profit (in millions)'], errors='coerce')
```

```python
print("\nDescriptive Statistics for Revenue (in millions):")
print(df_2["Revenue (in millions)"].describe())
```

```python
print("\nDescriptive Statistics for Profit (in millions):")
print(df_2["Profit (in millions)"].describe())
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_2["Revenue (in millions)"], df_2["Profit (in millions)"], alpha=0.7)
plt.title("Revenue vs. Profit")
plt.xlabel("Revenue (in millions)")
plt.ylabel("Profit (in millions)")
plt.show()
```

```python
correlation = df_2["Revenue (in millions)"].corr(df_2["Profit (in millions)"])
print(f"\nCorrelation between Revenue and Profit: {correlation:.2f}")
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_2["Revenue (in millions)"], bins=20, color='skyblue', edgecolor='black')
plt.title("Revenue Distribution (in millions)")
plt.xlabel("Revenue (in millions)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_2["Revenue (in millions)"], df_2["Profit (in millions)"], alpha=0.7)
plt.title("Revenue vs. Profit")
plt.xlabel("Revenue (in millions)")
plt.ylabel("Profit (in millions)")
plt.show()
```

```python
avg_revenue_by_year = df_2.groupby("Year")["Revenue (in millions)"].mean().reset_index()
print("\nAverage Revenue by Year:")
print(avg_revenue_by_year)
```

```python
top_companies = df_2.sort_values(by="Rank").head(30)
print("\nTop 30 Companies by Rank:")
print(top_companies)
```

```python
revenue_mean   = df_2['Revenue (in millions)'].mean()
revenue_median = df_2['Revenue (in millions)'].median()
revenue_std    = df_2['Revenue (in millions)'].std()
revenue_min    = df_2['Revenue (in millions)'].min()
revenue_max    = df_2['Revenue (in millions)'].max()
revenue_range  = revenue_max - revenue_min
revenue_skew   = df_2['Revenue (in millions)'].skew()
```

```python
print("Descriptive Statistics for Revenue (in millions):")
print(f"Mean: {revenue_mean:.2f}")
print(f"Median: {revenue_median:.2f}")
print(f"Standard Deviation: {revenue_std:.2f}")
print(f"Range: {revenue_range:.2f} (Min: {revenue_min:.2f}, Max: {revenue_max:.2f})")
print(f"Skewness: {revenue_skew:.2f}")
```

```python
if revenue_mean > revenue_median:
    print("\nInference: The revenue distribution may be right-skewed, as the mean is higher than the median.")
elif revenue_mean < revenue_median:
    print("\nInference: The revenue distribution may be left-skewed, as the mean is lower than the median.")
else:
    print("\nInference: The mean and median are similar, suggesting the distribution might be symmetric.")
```

```python
if revenue_std > 0.5 * revenue_mean:
    print("Inference: There is high variability in the revenue figures, which could indicate inconsistent performance across companies or time periods.")
else:
    print("Inference: The revenue figures show relatively low variability, indicating steadier performance.")
```

```python
print("\nObservation: The wide range from {0:.2f} to {1:.2f} may indicate the existence of outliers or a diverse set of revenue performances.".format(revenue_min, revenue_max))
```

```python
if revenue_skew > 0.5:
    print("Inference: Positive skewness indicates a tail on the right side; a few high revenue values might be pulling the mean up.")
elif revenue_skew < -0.5:
    print("Inference: Negative skewness indicates a tail on the left side; a few low revenue values might be lowering the mean.")
else:
    print("Inference: Skewness near zero suggests the distribution is fairly symmetric.")
```
