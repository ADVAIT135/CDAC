### 1. Comparison of Various Species Based on Sepal Length and Width

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', style='Species', s=80)
plt.title('Comparison of Iris Species: Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

#### The above scatter plot clearly shows the distribution of sepal dimensions for each species. Different markers and colors are used to make it easier to visually differentiate among Iris-setosa, Iris-versicolor, and Iris-virginica.

### 2. Bi-variate Analysis via Pairplot Visualization

```python
sns.pairplot(df, hue="Species", markers=["o", "s", "D"])
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

#### This pairplot not only illustrates pairwise relationships but also gives a glimpse into the univariate distributions of each feature. 
#### Some pairs (like petal dimensions) are extremely effective at separating the species, while others (such as sepal measurements) show partial overlap.

### 3. Justifying that Sepal Length and Sepal Width are Slightly Correlated

```python
correlation = df[['SepalLengthCm', 'SepalWidthCm']].corr().iloc[0,1]
print(f"Correlation between Sepal Length and Sepal Width: {correlation:.2f}")
```

```python
sns.lmplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', height=6, aspect=1.2, markers=["o", "s", "D"])
plt.title('Linear Relationship Between Sepal Length and Sepal Width')
plt.show()
```

#### The calculated correlation value gives a numeric justification.
#### The regression plot visually confirms that as one measurement increases, there is a tendency for the other to also increase, though they are not perfectly correlated—hence "slightly correlated."

### 4. Visualizing the Categorical Feature “Species” with All Other Four Input Variables

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

plt.figure(figsize=(16, 10))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='Species', y=feature, data=df)
    plt.title(f'{feature} Distribution by Species')
    plt.xlabel('Species')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

#### Iris-setosa often shows distinctly lower petal measurements compared to the other species.

### 5. Plotting the Density of the Length and Width in the Species

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.kdeplot(data=df, x="SepalLengthCm", hue="Species", fill=True, common_norm=False)
plt.title('Density Plot of Sepal Length by Species')
plt.xlabel('Sepal Length (cm)')

plt.subplot(1, 2, 2)
sns.kdeplot(data=df, x="SepalWidthCm", hue="Species", fill=True, common_norm=False)
plt.title('Density Plot of Sepal Width by Species')
plt.xlabel('Sepal Width (cm)')

plt.tight_layout()
plt.show()
```

### 6. Plotting the Probability Density Function (PDF) with Histogram and Corresponding Kernel Density Plot

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['PetalLengthCm'], kde=True, stat="density", bins=20, color='skyblue')
plt.title('Histogram and KDE of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['SepalLengthCm'], kde=True, stat="density", bins=20, color='red')
plt.title('Histogram and KDE of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['PetalWidthCm'], kde=True, stat="density", bins=20, color='green')
plt.title('Histogram and KDE of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['SepalWidthCm'], kde=True, stat="density", bins=20, color='brown')
plt.title('Histogram and KDE of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Density')
plt.show()
```

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
colors = ['red', 'brown', 'skyblue', 'green']

plt.figure(figsize=(16, 12))

for i, (feature, col) in enumerate(zip(features, colors), start=1):
    plt.subplot(2, 2, i)
    sns.histplot(df[feature], kde=True, stat="density", bins=20, color=col)
    plt.title(f'PDF of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```
