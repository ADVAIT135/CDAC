### 4. A.  Let's consider Premium cars to be those cars whose market category falls in 'LUXURY' class

```python

premium_cars_alt = data[data['Market Category'].str.contains("Luxury", case=False, na=False)]

premium_cars_alt = premium_cars_alt[['Make', 'Model', 'MSRP', 'Market Category']]
premium_cars_alt.to_excel('4 A. Premium Cars of each car company based on market category.xlsx', index = False)
```

### 4. B. Let's consider Premium cars to be those cars whose 'MSRP' is more than the AVG. MSRP of that particular company 'Make'

```python
df = data
```

```python
df['avg_MSRP'] = df.groupby('Make')['MSRP'].transform('mean')
```

```python
premium_cars = df[df['MSRP'] > df['avg_MSRP']]
```

```python
premium_cars_selected = premium_cars[['Make', 'Model', 'MSRP']]
```

```python
premium_cars_selected.to_excel('4 B. Premium Cars of each car company based on avg price.xlsx', index = False)
```

```python
premium_luxury_cars = pd.read_excel('4 B. Premium Cars of each car company based on avg price.xlsx')
```

## 5. Average price of cars each car company make.

```python
avg_price_by_make = df.groupby("Make")["MSRP"].mean().reset_index().rename(columns={"MSRP": "avg_price"})
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
avg_price_by_make
```
